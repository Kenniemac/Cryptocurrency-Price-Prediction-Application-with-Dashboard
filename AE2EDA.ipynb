```python
#data processing
import numpy as np
import pandas as pd
from datetime import date

#get time series data
import yfinance as yf

#Prophet model for time series forecast
from prophet import Prophet
from prophet.plot import add_changepoints_to_plot, plot_cross_validation_metric
from prophet.diagnostics import cross_validation, performance_metrics

#Visualization
import plotly.graph_objs as go


df = pd.DataFrame()


#data extraction for 20 cryptocoins
crypto_symbols = ['BTC','ETH','LTC','DOGE','SOL','USDT','XRP','ADA','XLM','BNB','DOT','DASH','BSV','VET','AVAX','EOS','ZEC','CEL','HEX','TRX']
cryptocurrencies = ['Bitcoin','Ethereum','Litecoin','Dogecoin','Solana','Tether','XRP','Cardano','Stellar','Binance','Polkadot','Dash','BitcoinSV',
                    'VeChain','Avalanche','EOSCoin','ZCash','Celsius','HEX','TRON']

for coins in crypto_symbols:
    coin_index = crypto_symbols.index(coins)

    yf_data = yf.Ticker(f'{coins}-USD').history(start='2020-01-01', end=date.today().strftime("%Y-%m-%d"), interval='1d')


    cdf = pd.DataFrame(yf_data)
    cdf['name'] = cryptocurrencies[coin_index]
    cdf['symbol'] = coins
    df = df.append(cdf)



df.reset_index(inplace = True)

#this is used to set the date as index
df['Date'] = pd.to_datetime(df['Date'])

df.head(10)
```

```python
from datetime import datetime, timedelta
# Specify a cutoff day range.
days = 60

# Create a cutoff date using the days value.
today = datetime.now()
stop_date = today - timedelta( days )

# Use cutoff_date to split our dataset to history and actuals, which we will use to validate the model.
df_train = df_data[df_data['ds'] <= stop_date]
df_test = df_data[df_data['ds'] > stop_date]


#print the number of train and test data
print("The training dataset has", len(df_train))
print("The test dataset has", len(df_test))
```

```python
from prophet import Prophet

#create the prophet model with confidence internal of 95%
m = Prophet(changepoint_prior_scale=0.01, seasonality_prior_scale=0.01)

#fit the model using the training dataset
m.fit(df_train)
```

```python
#create a future dataframe for prediction
future = m.make_future_dataframe(periods=31)

#forecast the future dataframe
forecast = m.predict(future)

#check the predicted values and upper and lower bound
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
#visualize the forecast
fig = m.plot(forecast)
ax = fig.gca()

ax.plot(df_test["ds"], df_test["y"], 'r.')
```
